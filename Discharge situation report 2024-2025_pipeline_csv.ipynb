```python
import pandas as pd
import os

folder_path = r"C:\Users\e410675\OneDrive - London South Bank University\LSBU - Jose's Collaborative Projects\Norfolk and Waveney Demand Modelling\New Data\Discharge situation report 2024-2025"
combined_data = []

for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        try:
            data = pd.read_csv(file_path, encoding='windows-1252')  # Change encoding here
            data['source file name'] = file_name
            combined_data.append(data)
        except Exception as e:
            print(f"Failed to read {file_name}: {e}")


# Concatenate all dataframes
combined_data = pd.concat(combined_data, axis=0, ignore_index=True)

df=combined_data.copy()
```

```python
#Rmove the column  Include?
df.drop(columns=["Include?"], inplace=True)

#Replace Weekly snaphot metric for Weekly snapshot metric
df["Metric Type"] = df["Metric Type"].replace("Weekly snaphot metric", "Weekly snapshot metric")
```
