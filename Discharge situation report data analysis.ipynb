# Data Cleaning

```python
import pandas as pd


# Function to properly capitalize each word, except "NHS"
def format_organisation_name(name):
    words = name.split()
    formatted_words = [word.upper() if word.upper() == "NHS" else word.capitalize() for word in words]
    return " ".join(formatted_words)

# Apply the formatting function
df["Org Name"] = df["Org Name"].apply(format_organisation_name)
```

```python
#The metric Capacity – Awaiting restart of existing social care arrangements (Pathway 0) has duplicates for each month. Remove these duplicates.

# Check for duplicates across all columns
duplicate_rows = df[df.duplicated()]
print(f"Duplicate rows found: {len(duplicate_rows)}")

# Drop exact duplicates
df_cleaned = df.drop_duplicates()

# Optional: only drop based on a subset of identifying columns
df_cleaned = df.drop_duplicates(subset=[
    'Period', 'Org Code', 'Metric', 'Value'
])

df=df_cleaned.copy()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os


# === Define target organisations ===
target_orgs = [
    "Dartford And Gravesham NHS Trust",
    "Maidstone And Tunbridge Wells NHS Trust",
    "Medway NHS Foundation Trust",
    "East Kent Hospitals University NHS Foundation Trust"
]

# === Filter for Discharge destination and target organisations ===
discharge_df = df[
    (df["Metric Group"] == "Discharge destination") & 
    (df["Org Name"].isin(target_orgs))
].copy()

# === Convert 'Value' to numeric ===
discharge_df["Value"] = pd.to_numeric(discharge_df["Value"], errors="coerce")

# === Clean 'Metric' to extract destination ===
discharge_df["Discharge Destination"] = (
    discharge_df["Metric"]
    .str.replace("\xa0", " ", regex=False)
    .str.strip()
)

# === Group by Discharge Destination and Organisation ===
agg_discharge = (
    discharge_df.groupby(["Discharge Destination", "Org Name"])["Value"]
    .sum()
    .reset_index()
    .sort_values("Value", ascending=True)
)

# === Set Org Name as categorical with target_orgs order ===
agg_discharge["Org Name"] = pd.Categorical(agg_discharge["Org Name"], categories=target_orgs, ordered=True)

# === Function to break long labels into two lines ===
def break_long_labels(label, max_length=30):
    if len(label) > max_length:
        parts = label.split(" ")
        midpoint = len(parts) // 2
        return " ".join(parts[:midpoint]) + "\n" + " ".join(parts[midpoint:])
    return label

# === Apply label shortening ===
agg_discharge["Discharge Destination"] = agg_discharge["Discharge Destination"].apply(break_long_labels)

# === Order discharge destinations by total value across all orgs ===
destination_order = (
    agg_discharge.groupby("Discharge Destination")["Value"]
    .sum()
    .sort_values(ascending=True)
    .index
)

agg_discharge["Discharge Destination"] = pd.Categorical(
    agg_discharge["Discharge Destination"],
    categories=destination_order,
    ordered=True
)

# === Plot ===
fig, ax = plt.subplots(figsize=(20, 8))  # Increased width to use space better
fig.patch.set_facecolor('white')

sns.barplot(
    data=agg_discharge,
    y="Discharge Destination",
    x="Value",
    hue="Org Name",
    hue_order=target_orgs,
    dodge=True,
    ax=ax
)

# === Labels and formatting ===
ax.set_xlabel("Number of Patients", fontsize=14)
ax.set_ylabel("Discharge Destination", fontsize=14)
ax.tick_params(axis='x', labelsize=13)
ax.tick_params(axis='y', labelsize=13)

# === Title and subtitle ===
ax.set_title(
    "Discharge Pathways for Discharged Patients (June 2024 – April 2025)",
    fontsize=20,
    pad=40,
    weight='bold',
    loc='center'
)
ax.text(
    0.5, 1.03,
    "Total Number of Discharges by Destination across Selected NHS Trusts",
    fontsize=16,
    ha='center',
    transform=ax.transAxes
)

# === Grid and style ===
ax.grid(axis='x', linestyle=':', linewidth=1, color='lightgrey', alpha=0.8)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# === Legend ===
ax.legend(
    fontsize=12,
    title_fontsize=12,
    loc='upper center',
    bbox_to_anchor=(0.5, -0.12),
    ncol=2,
    frameon=False
)

# === Final layout adjustment ===
plt.subplots_adjust(left=0.25, right=0.95, top=0.9, bottom=0.2)
plt.tight_layout(rect=[0, 0.1, 1, 1])

# === Save & show ===
plt.savefig(os.path.join(plot_save_path, "Discharge Pathways.png"), bbox_inches='tight', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import seaborn as sns
import os
from matplotlib.font_manager import FontProperties

# === Define metrics ===
metric_map = {
    "Number of patients discharged": "Discharged",
    "Number of patients remaining in hospital who no longer meet the criteria to reside": "Remaining in Hospital"
}

# === Clean data ===
filtered_df = df[df["Metric"].isin(metric_map.keys())].copy()
filtered_df["Month"] = pd.to_datetime(filtered_df["Period"], format="%d/%m/%Y", errors='coerce').dt.to_period("M")
filtered_df["Month"] = filtered_df["Month"].dt.to_timestamp()
filtered_df["Value"] = pd.to_numeric(filtered_df["Value"], errors="coerce")
filtered_df = filtered_df[filtered_df["Value"] != 0]  # ✅ Filter out zero values
filtered_df["Metric Label"] = filtered_df["Metric"].map(metric_map)

# === Define Trusts ===
target_orgs = [
    "Dartford And Gravesham NHS Trust",
    "Maidstone And Tunbridge Wells NHS Trust",
    "East Kent Hospitals University NHS Foundation Trust"
]

# === Colors ===
colors = sns.color_palette("tab10", n_colors=2)

# === Plot per Trust ===
for org in target_orgs:
    org_df = filtered_df[filtered_df["Org Name"] == org].copy()
    if org_df.empty:
        continue

    pivot_df = (
        org_df.pivot_table(index="Month", columns="Metric Label", values="Value", aggfunc="sum")
        .fillna(0)
        .sort_index()
    )

    for col in metric_map.values():
        if col not in pivot_df.columns:
            pivot_df[col] = 0

    # Skip plot if all values are zero
    if pivot_df.sum().sum() == 0:
        continue

    # === Plot ===
    fig, ax = plt.subplots(figsize=(16, 8))
    fig.patch.set_facecolor('white')
    x_vals = mdates.date2num(pivot_df.index.to_pydatetime())
    bottom = pd.Series([0.0] * len(pivot_df), index=pivot_df.index)

    # Plot in reverse order: Remaining (bottom), Discharged (top)
    for i, col in enumerate(reversed(list(metric_map.values()))):
        y1 = bottom.values
        y2 = (bottom + pivot_df[col]).values
        ax.fill_between(x_vals, y1, y2, color=colors[i], label=col)
        bottom += pivot_df[col]

    # === Titles and subtitle ===
    ax.set_title(
        "Outcomes For Patients Who No Longer Meet The Criteria To Reside",
        fontsize=20,
        pad=40,
        weight='bold',
        loc='center'
    )
    ax.text(
        0.5, 1.03,
        org,
        fontsize=16,
        ha='center',
        transform=ax.transAxes
    )

    # === Axis formatting ===
    ax.set_xlabel("Year-Month", fontsize=14)
    ax.set_ylabel("Number of Patients", fontsize=14)
    ax.tick_params(axis='x', labelsize=13, rotation=0)
    ax.tick_params(axis='y', labelsize=13)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{int(x):,}"))

    # === X-axis ticks ===
    labels = pivot_df.index.strftime("%Y-%m").tolist()
    display_labels = labels if len(labels) <= 13 else [label if i % 2 == 0 else '' for i, label in enumerate(labels)]
    ax.set_xticks(x_vals)
    ax.set_xticklabels(display_labels, ha='center')

    # === Grid and borders ===
    ax.grid(True, linestyle=':', linewidth=1, color='lightgrey', alpha=0.8)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # === Font configuration ===
    bold_font = FontProperties()
    bold_font.set_weight('bold')

    # === Legend ===
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(
        handles, labels,
        title_fontproperties=bold_font,
        fontsize=12,
        loc='upper center',
        bbox_to_anchor=(0.5, -0.1),
        ncol=2,
        frameon=False
    )

    # === Final layout ===
    plt.tight_layout(rect=[0, 0.1, 1, 1])
    plt.savefig(os.path.join(plot_save_path, f"o_Patients_Who_No_Longer_Meet_The_Criteria_To_Reside_{org}.png"), bbox_inches='tight', dpi=300)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
import os

# === Define target organisations ===
target_orgs = [
    "Dartford And Gravesham NHS Trust",
    "Maidstone And Tunbridge Wells NHS Trust",
    "Medway NHS Foundation Trust",
    "East Kent Hospitals University NHS Foundation Trust"
]

# === Filter data ===
filtered_exact_df = df[
    (df["Org Name"].isin(target_orgs)) &
    (df["Metric"] == "Number of patients discharged")
].copy()

# === Convert columns ===
filtered_exact_df["Period"] = pd.to_datetime(filtered_exact_df["Period"], format="%d/%m/%Y", errors='coerce')
filtered_exact_df["Value"] = pd.to_numeric(filtered_exact_df["Value"], errors="coerce")
filtered_exact_df["Month"] = filtered_exact_df["Period"].dt.to_period("M")

# === Aggregate to monthly totals ===
monthly_totals_df = (
    filtered_exact_df
    .groupby(["Org Name", "Month"], as_index=False)["Value"]
    .sum()
)

# === Remove rows with Value == 0 ===
monthly_totals_df = monthly_totals_df[monthly_totals_df["Value"] != 0]

# === Convert month back to datetime for plotting ===
monthly_totals_df["Month"] = monthly_totals_df["Month"].dt.to_timestamp()

# === Plot with style ===
fig, ax = plt.subplots(figsize=(16, 8))
fig.patch.set_facecolor('white')
tab10 = plt.get_cmap('tab10')

for idx, org in enumerate(target_orgs):
    org_data = monthly_totals_df[monthly_totals_df["Org Name"] == org]
    ax.plot(
        org_data["Month"],
        org_data["Value"],
        label=org,
        color=tab10(idx),
        linewidth=2,
        linestyle='-',
        marker='o'
    )

# === Y-axis formatting ===
ax.set_ylabel("Number Of Patients", fontsize=14)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{int(x):,}"))
ax.tick_params(axis='y', labelsize=13)
ax.set_ylim(bottom=0)

# === X-axis formatting ===
ax.set_xlabel("Year-Month", fontsize=14)
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
ax.tick_params(axis='x', rotation=0, labelsize=13)

# === Titles ===
ax.set_title(
    "Number Of Patients Discharged",
    fontsize=20,
    pad=40,
    weight='bold',
    loc='center'
)
ax.text(
    0.5, 1.03,
    "Comparison Across Selected NHS Foundation Trusts",
    fontsize=16,
    ha='center',
    transform=ax.transAxes
)

# === Grid and style ===
ax.grid(True, linestyle=':', linewidth=1, color='lightgrey', alpha=0.8)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# === Legend ===
plt.legend(
    fontsize=12,
    loc='upper center',
    bbox_to_anchor=(0.5, -0.1),
    ncol=2,
    frameon=False
)

# === Final layout ===
plt.tight_layout(rect=[0, 0.1, 1, 1])

# === Save & show ===
plt.savefig(os.path.join(plot_save_path, "patients-Number of patients discharged.png"), bbox_inches='tight', dpi=300)
plt.show()
```
